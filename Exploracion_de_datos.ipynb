Leemos los archivos de ondas sanas.

```python
! unzip /content/Normal1.zip?raw=true
readerN=pd.read_csv("Normal1.csv")
xN=list(np.transpose(np.array(readerN["0"])))
```

Miramos la cantidad de vectores que exportamos sanos y graficamos 1.
Como sabemos que las ondas tienen frecuencia 60 Hz y son 10 oscilaciones hacemos el calculo respectivo para determinar la longitud de este y asi graficarlo de manera correcta.

```python
print("La cantidad de vectores sanos es:",len(list(readerN)))
print("La cantidad de puntos en un vector es :",len(xN))
def reconstruction_plot(yyy, **kwargs):
    """Plot signal vector on x [0,1] independently of amount of values it contains."""
    plt.plot(np.linspace(0, (1/533)*(10/60)* len(yyy), len(yyy)), yyy, **kwargs)
reconstruction_plot(xN) # Original 
plt.title("Onda normal")
plt.show()
```

Hacemos lo mismo para las ondas distorsionadas , pero graficaremos 2 cualquiera  a ver si son distinas

```python
! unzip /content/sag2.zip?raw=true
readers=pd.read_csv("sag2.csv")
print("La cantidad de vectores con sags es:",len(list(readers)))
x=list(np.transpose(np.array(readers[str(rd.randint(0,1000))])))
x2=list(np.transpose(np.array(readers[str(rd.randint(0,1000))])))
print("La cantidad de puntos en un sag es:",len(x))

reconstruction_plot(x)
reconstruction_plot(x2)
plt.title("Onda con sag")
plt.show()
```

Notese que las ondas presentan la distorsion en diferentes momentos, todas duran 10 ciclos ,una cantidad fija de puntos y estan normalizadas
